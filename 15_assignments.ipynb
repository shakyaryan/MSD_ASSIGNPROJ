7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...

Answer:

```python
def genPrimes():
    primes = []
    num = 2
    while True:
        for p in primes:
            if num % p == 0:
                break
        else:
            primes.append(num)
            yield num
        num += 1
```

Example usage:

```python
g = genPrimes()
print(next(g))  # 2
print(next(g))  # 3
print(next(g))  # 5
print(next(g))  # 7
```
