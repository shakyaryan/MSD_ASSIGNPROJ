### **1. Is the Python Standard Library included with PyInputPlus?**

No, PyInputPlus is not part of the Python Standard Library.
It is a third-party module that needs to be installed separately using the command `pip install pyinputplus`.
The Python Standard Library includes built-in modules, but PyInputPlus is not one of them.

### **2. Why is PyInputPlus commonly imported with `import pyinputplus as pypi`?**

PyInputPlus is commonly imported as `pypi` to make it easier and shorter to use in programs.
Instead of writing `pyinputplus.inputInt()`, we can simply write `pypi.inputInt()`.
It saves typing and makes the code look cleaner.

### **3. How do you distinguish between `inputInt()` and `inputFloat()`?**

The `inputInt()` function allows only whole numbers (integers) as input.
The `inputFloat()` function allows decimal numbers as input.
For example, entering `10` works for both, but entering `10.5` will only work with `inputFloat()` and not with `inputInt()`.

### **4. Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?**

You can use the `inputInt()` function with `min` and `max` arguments like this:
`pyinputplus.inputInt(min=0, max=99)`
This makes sure that the user enters an integer value and that it lies between 0 and 99.

### **5. What is transferred to the keyword arguments `allowRegexes` and `blockRegexes`?**

The `allowRegexes` argument is used to specify a list of regular expressions that are allowed as valid input.
The `blockRegexes` argument is used to specify a list of regular expressions that should be rejected as invalid input.
They help in controlling what kind of text inputs are accepted or blocked.

### **6. If a blank input is entered three times, what does `inputStr(limit=3)` do?**

If the user enters a blank input three times, the function stops asking for input and raises a `RetryLimitException` error.
This happens because the limit of 3 tries has been reached.

### **7. If a blank input is entered three times, what does `inputStr(limit=3, default='hello')` do?**

If the user enters blank input three times, the function will not show an error.
Instead, it will return the default value `'hello'` after the three failed attempts.
So, the program continues smoothly using `'hello'` as the input.
