
### **1. Add the current date to the text file today.txt as a string.**

You open or create the file **today.txt** and write today’s date into it as plain text (for example “2025-11-13”). This saves the current date inside the file.

---

### **2. Read the text file today.txt into the string today_string.**

You open **today.txt**, read the content inside it, and store that text in a variable called **today_string**. This variable now contains the date you wrote earlier.

---

### **3. Parse the date from today_string.**

You take the text stored in **today_string** and convert it into a proper date object. This means you turn the plain text date into a format that can be used for calculations or comparisons.

---

### **4. List the files in your current directory.**

You request a list of all items in the folder you are currently working in. This will show all files and subfolders present in that directory.

---

### **5. Create a list of all of the files in your parent directory (minimum five files should be available).**

You move one level up to the parent directory and collect the names of all files there into a list. There should be at least five files available in that folder, and all their names are stored in the list.

---

### **6. Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between one and five, print the current time, and then exit.**

You create three independent worker processes. Each process pauses for a randomly chosen time between 1 and 5 seconds, then prints the current time, and then stops. All three run at the same time, so their outputs appear in random order.

---

### **7. Create a date object of your day of birth.**

You make a date object using your birth year, month, and day. This object represents your exact birth date in a structured format.

---

### **8. What day of the week was your day of birth?**

You check the weekday of your date-of-birth object. It tells you whether you were born on Monday, Tuesday, Wednesday, etc.

---

### **9. When will you be (or when were you) 10,000 days old?**

You add **10,000 days** to your birth date. The resulting date is the exact day you will become (or already became) 10,000 days old.
