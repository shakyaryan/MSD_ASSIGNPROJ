1.

The code prints iNeuron.
Reason: The function only reads the value of X. Since there is no X inside the function, Python uses the global X, which contains the string iNeuron.

---

2.

The final output is iNeuron.
Reason: Inside the function a new local variable X is created with the value NI!. It does not change the global X. After calling the function, printing X still shows the original global value iNeuron.

---

3.

The program prints two lines:
First it prints NI.
Then it prints iNeuron.
Reason: Inside the function, X is a local variable containing NI, so that is printed first. The global X is still iNeuron and has not changed, so the print after the function prints iNeuron.

---

4.

The program prints NI.
Reason: The function uses the global statement, which means the assignment changes the global X. After the function runs, the global X becomes NI, so printing X shows NI.

---

5.

The function prints NI.
The global X remains iNeuron.
Reason: Inside func, X is a local variable with the value NI. The nested function does not have its own X, so it reads the X from the enclosing function. It prints NI. This does not affect the global X.

---

6.

The program prints Spam.
Reason: The variable X in func initially has the value NI. Inside nested, the nonlocal statement tells Python to use the X from the outer function. The nested function changes that X to Spam. Therefore, when func prints X, the value is Spam.
