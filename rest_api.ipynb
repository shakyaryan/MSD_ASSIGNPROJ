

**1. What is a RESTful API?**
A RESTful API is a web service that follows REST architecture and uses HTTP methods to perform operations on resources using URLs.

---

**2. What is Flask, and why is it popular for building APIs?**
Flask is a lightweight Python web framework. It is popular because it is simple, flexible, and easy to build REST APIs quickly.

---

**3. What are HTTP methods used in RESTful APIs?**
Common HTTP methods:
GET → Retrieve data
POST → Create new resource
PUT → Update resource (full update)
PATCH → Partial update
DELETE → Remove resource

---

**4. What is the purpose of the @app.route() decorator in Flask?**
@app.route() maps a URL to a Python function (view function) that runs when that URL is accessed.

---

**5. What is the role of Flask-SQLAlchemy?**
Flask-SQLAlchemy is used to connect Flask with databases and perform ORM-based database operations.

---

**6. How do you create a basic Flask application?**
Create Flask object → Define routes → Run the app.

Example:

```python
from flask import Flask
app = Flask(__name__)

@app.route('/')
def home():
    return "Hello"

app.run(debug=True)
```

---

**7. How do you return JSON responses in Flask?**
Use jsonify() to send JSON response.

```python
from flask import jsonify
return jsonify({"message":"Success"})
```

---

**8. How do you handle POST requests in Flask?**
Use methods=['POST'] and request.get_json().

```python
from flask import request

@app.route('/add', methods=['POST'])
def add():
    data = request.get_json()
    return jsonify(data)
```

---

**9. How do you handle errors in Flask (e.g., 404)?**
Use error handlers.

```python
@app.errorhandler(404)
def not_found(e):
    return "Page not found", 404
```

---

**10. How do you structure a Flask app using Blueprints?**
Blueprints split a large app into modules (routes, views, APIs).
They help organize large projects and keep code clean.
