1. Create a zoo.py file first. Define the hours() function, which prints the string 'Open 9-5 daily'. Then, use the interactive interpreter to import the zoo module and call its hours() function.

Answer:

Create a file named zoo.py and define a function called hours() that prints “Open 9-5 daily”.
After that, open the Python interactive interpreter, import the zoo module, and call the function using zoo.hours().

2. In the interactive interpreter, import the zoo module as menagerie and call its hours() function.

Answer:

In the interpreter, import the module using an alias with the statement import zoo as menagerie, and then call the function using menagerie.hours().

3. Using the interpreter, explicitly import and call the hours() function from zoo.

Answer:

Use the statement from zoo import hours to import only the function, and then call it directly by typing hours().

4. Import the hours() function as info and call it.

Answer:

Use the statement from zoo import hours as info to import the function with a new name, and then call it using info().

5. Create a plain dictionary with the key-value pairs 'a': 1, 'b': 2, and 'c': 3, and print it out.

Answer:

Create a dictionary named plain containing the key-value pairs a:1, b:2, and c:3. Then print it to display its contents.

6. Make an OrderedDict called fancy from the same pairs listed in 5 and print it. Did it print in the same order as plain?

Answer:

Create an OrderedDict named fancy using the same key-value pairs. When printed, it shows the same order as the plain dictionary because modern Python dictionaries also preserve insertion order.

7. Make a default dictionary called dict_of_lists and pass it the argument list. Make the list dict_of_lists['a'] and append the value 'something for a' to it in one assignment. Print dict_of_lists['a'].

Answer:

Create a defaultdict object named dict_of_lists and pass list as the default factory function. Append the string “something for a” to dict_of_lists['a'] and then print it. It will display ['something for a'].
