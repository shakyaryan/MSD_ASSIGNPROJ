```python
#   1. Write a Python program to find words which are greater than given length k?

# Input sentence
text = input("Enter a sentence: ")

# Input k (length)
k = int(input("Enter the length k: "))

# Split sentence into words
words = text.split()

# Empty list to store result
result = []

# Check each word
for word in words:
    if len(word) > k:
        result.append(word)

# Output the result
print("Words greater than length", k, "are:", result)
```

```python
# ques 2
# Removing i-th character from a string

string = input("Enter a string: ")
i = int(input("Enter index to remove: "))

# String slicing
new_string = string[:i] + string[i+1:]

print("After removing character:", new_string)
```

```python
# ques3
# Split and Join a string

text = input("Enter a sentence: ")

# Split into words
words = text.split()

# Join using '-'
joined_string = "-".join(words)

print("Split words:", words)
print("Joined string:", joined_string)
```

```python
# ques 4
# Check if string is binary

s = input("Enter a string: ")

is_binary = True

for ch in s:
    if ch not in ('0', '1'):
        is_binary = False
        break

if is_binary:
    print("Binary String")
else:
    print("Not a Binary String")
```

```python
#ques5
# Uncommon words from two strings

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

w1 = s1.split()
w2 = s2.split()

uncommon = []

for word in w1:
    if word not in w2:
        uncommon.append(word)

for word in w2:
    if word not in w1:
        uncommon.append(word)

print("Uncommon words are:", uncommon)
```

```python
# ques6
# Find duplicate characters

s = input("Enter a string: ")

duplicates = []

for ch in s:
    if s.count(ch) > 1 and ch not in duplicates:
        duplicates.append(ch)

print("Duplicate characters:", duplicates)
```

```python
# ques7
# Check special character in string

s = input("Enter a string: ")

special = False

for ch in s:
    if not ch.isalnum():   # not letter or number
        special = True
        break

if special:
    print("String contains special characters.")
else:
    print("No special characters found.")
```
