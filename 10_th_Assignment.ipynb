
**1. Difference between shutil.copy() and shutil.copytree():**
The function shutil.copy() is used to copy a single file from one location to another.
The function shutil.copytree() is used to copy an entire folder (including all its files and subfolders) to a new location.

---

**2. Function used to rename files:**
The function used to rename files is os.rename().
It can also move a file from one folder to another.

---

**3. Difference between delete functions in send2trash and shutil modules:**
The send2trash module sends the file or folder to the Recycle Bin, so it can be restored later.
The shutil module permanently deletes the file or folder, and it cannot be recovered.

---

**4. ZipFile method equivalent to File objects’ open() method:**
The ZipFile() function from the zipfile module is equivalent to the open() method of file objects.
It is used to open and work with ZIP files.

---

**5. Program to search a folder tree for files with a certain extension and copy them to a new folder:**
We can write a program that goes through all folders and subfolders using os.walk().
It checks each file’s extension (like .pdf or .jpg) and copies the matching files to a new folder using shutil.copy().

Example:

```python
import os, shutil

source = "C:\\Users\\Example\\Documents"
destination = "C:\\Users\\Example\\Collected_Files"
extension = ".pdf"

for folder, subfolders, files in os.walk(source):
    for file in files:
        if file.endswith(extension):
            shutil.copy(os.path.join(folder, file), destination)
