1.

Def statements and lambda expressions both create functions. A def statement gives a function a name and a block of statements; it can contain multiple statements, docstrings, and assigns the function to a name. A lambda expression creates an anonymous, single-expression function and is usually used where a small one-line function is needed temporarily.

2.

The benefit of lambda is brevity and convenience for small throwaway functions. Lambdas let you define simple behaviour inline (for example as an argument to another function) without writing a full def and a separate name. They keep short functions close to where they are used.

3.

Map, filter, and reduce are tools for processing collections. Map applies a function to every item in a sequence and returns all the results. Filter applies a predicate (a function that returns true or false) to each item and returns only the items where the predicate is true. Reduce takes a binary function and cumulatively applies it to the items of a sequence to produce a single result (for example summing all items). Map transforms every element, filter selects elements, and reduce aggregates elements into one value.

4.

Function annotations are optional metadata added to function parameters and to the return value. They are written next to parameter names and after the arrow for the return, and they do not change program behavior by themselves. Annotations are used for documentation, type hints, or by external tools (linters, type checkers, IDEs, frameworks) to validate or analyze code.

5.

Recursive functions are functions that call themselves, directly or indirectly. They are used to solve problems that can be broken into smaller instances of the same problem (for example traversing tree structures, computing factorials, or implementing divide-and-conquer algorithms). Each recursive call works on a smaller piece of the problem and there must be a base case that stops further recursion to avoid infinite recursion.

6.

General design guidelines for functions:
• Keep functions small and focused: each function should do one clear thing.
• Give functions descriptive names that reflect their job.
• Prefer clear arguments and avoid too many parameters; group related parameters into objects if needed.
• Avoid side effects when possible; functions that return values and do not modify external state are easier to test and reuse.
• Document behavior, inputs, outputs, and exceptions (use docstrings).
• Use consistent argument ordering and sensible defaults.
• Handle errors explicitly and fail fast with helpful messages.
• Write pure logic where practical so functions are predictable and testable.

7.

Ways functions can communicate results to a caller:
• Return a value (or tuple of values) using the return statement.
• Modify and return a mutable argument (for example change a list or dict passed in).
• Raise an exception to signal an error condition (the caller can catch it and react).
• Write to external outputs or state that the caller can observe (for example write to a file, database, or a global variable).
• Use callbacks or pass in a function/object that the called function invokes with results.
• Yield values via a generator so the caller can iterate over results lazily.
