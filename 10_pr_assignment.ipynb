```python
#   1. Write a Python program to find sum of elements in list?
l=[1,2,3,4,5,6]
sum =0
for i in l:
    sum=sum+i
print(sum)
```

```python
# 3. Write a Python program to find smallest number in a list?

l = [1, 2, 3, 4, 5, 6]

smallest = l[0]   # assume first element is smallest

for num in l:
    if num < smallest:
        smallest = num

print("Smallest number:", smallest)
```

```python
#   4. Write a Python program to find largest number in a list?
l=[23,354,5,6678,789,8]
largest=l[0] # assume first element is greater

for num in l:
    if num > largest:
        largest= num
    
print("Greater number :" , largest )
```

```python
#   5. Write a Python program to find second largest number in a list?


l=[23,354,5,6678,789,8]
largest=l[0] # assume first element is greater

second_largest = float("-inf") # very small number

for num in l:
    if num>largest:
        second_largest=largest
        largest = num
    elif num > second_largest and num!= largest :
        second_largest=num

print("Second largest number is :" , second_largest)
```

```python
#   6. Write a Python program to find N largest elements from a list?

l = [10 , 4, 20 , 15 , 2, 30]

largest = float("-inf")
second_largest = float("-inf")
third_largest = float("-inf")

for num in l:

    # 1st condition : num becomes largest
    if num > largest :
        third_largest = second_largest
        second_largest= largest
        largest = num

        # 2nd condtion: num becomes second largest
    elif num> second_largest:
        third_largest = second_largest
        second_largest = num

        #3rd condition num becomes third largest
    elif num> third_largest:
        third_largest= num

print("Top 3 largest number : ")
print(largest , second_largest , third_largest)
```

```python
# 8. Write a Python program to print odd numbers in a List?

l = [10 ,33,333,33333,33333, 4, 20 , 15 , 2, 30]
for i in l:
      if i % 2 != 0:
            print("odd number : ", i)
```

```python
# 9. Write a Python program to Remove empty List from List?

my_list = [1, [], 3, [], 5, [], 7]

new_list = []
for item in my_list:
    if item != []:
        new_list.append(item)

print(new_list)

```
