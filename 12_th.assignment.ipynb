1. In what modes should the PdfFileReader() and PdfFileWriter() File objects be opened?

Answer:

PdfFileReader() objects should be opened in 'rb' (read binary) mode, and PdfFileWriter() objects should be opened in 'wb' (write binary) mode.

2. From a PdfFileReader object, how do you get a Page object for page 5?

Answer:

pageObj = pdfReader.getPage(4)

(Page numbering starts from 0, so page 5 is page 4.)

3. What PdfFileReader variable stores the number of pages in the PDF document?

Answer:

The variable pdfReader.numPages stores the number of pages in the PDF document.

4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do before you can obtain Page objects from it?

Answer:

You must call pdfReader.decrypt('swordfish') before accessing the pages.

5. What methods do you use to rotate a page?

Answer:

You can use pageObj.rotateClockwise(degrees) or pageObj.rotateCounterClockwise(degrees).

6. What is the difference between a Run object and a Paragraph object?

Answer:

A Paragraph object represents an entire paragraph of text, while a Run object represents a section of text within that paragraph having the same style or formatting.

7. How do you obtain a list of Paragraph objects for a Document object that’s stored in a variable named doc?

Answer:

You can use doc.paragraphs to obtain a list of Paragraph objects.

8. What type of object has bold, underline, italic, strike, and outline variables?

Answer:

A Run object has bold, underline, italic, strike, and outline variables.

9. What is the difference between False, True, and None for the bold variable?

Answer:

False means the text is not bold, True means the text is bold, and None means the text inherits the bold setting from the document’s default style.

10. How do you create a Document object for a new Word document?

Answer:

You can create it using doc = docx.Document().

11. How do you add a paragraph with the text 'Hello, there!' to a Document object stored in a variable named doc?

Answer:

You can use doc.add_paragraph('Hello, there!').

12. What integers represent the levels of headings available in Word documents?

Answer:

The integers 0 through 9 represent the levels of headings, with 0 being the title and 1–9 being heading levels.
