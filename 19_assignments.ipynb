1. Create a class called Thing with no contents and print it. Then, create an object called example from this class and also print it. Are the printed values the same or different?

Answer:
A class named Thing is created using the pass statement. When the class itself is printed, it displays information about the class definition. When an object is created from the class and printed, it shows the object’s memory location. Therefore, the printed values are different — one represents the class, and the other represents the object instance.

2. Create a new class called Thing2 and add the value 'abc' to the letters class attribute. Letters should be printed.

Answer:
A class Thing2 is created with a class attribute named letters containing the value 'abc'. When the class attribute is printed, it displays 'abc'. Class attributes belong to the class itself and can be accessed without creating an object.

3. Make yet another class called Thing3. This time, assign the value 'xyz' to an instance (object) attribute called letters. Print letters. Do you need to make an object from the class to do this?

Answer:
In the class Thing3, the attribute letters is defined as an instance attribute. This means that each object created from the class has its own copy of the attribute. To access and print the value of letters, an object must first be created from the class. Yes, an object is required to access instance attributes.

8. Modify Element to make the attributes name, symbol, and number private. Define a getter property for each to return its value.

Answer:
