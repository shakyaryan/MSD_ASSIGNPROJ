```python
# Write a Python Program to Add Two Matrices?

A =[
    [1,2,3],
    [4,5,6]
]

B= [
    [7,8,9],
    [1,2,3]
    ]

result = [
    [0,0,0],
    [0,0,0]
]

# add matrices
for i in range(2): # row 
    for j in range(3):# column
        result[i][j] = A[i][j]+B[i][j]

# print result in matrix form
print("Resultant Matrix :")
for i in range(2):
    for j in range(3):
        print(result[i][j] , end=' ')
    print()
    
```

```python
# Write a Python Program to Multiply Two Matrices?
A =[
    [1,2,3],
    [4,5,6]
]

B= [
    [7,8],
    [1,2],
    [1,1]
    ]

result = [
    [0,0],
    [0,0]
]

# matrix multiplication
for i in range(2): # rows of A
    for j in range(2): # columns of B
        for k in range(3):# columns of a/rows of b
            result[i][j] += A[i][k]*B[k][j]

# print reuslt

print("Resultant Matrix:") 
for i in range(2):
    for j in range(2):
        print(result[i][j] , end = " ")
    print()

```

```python
# Write a Python Program to Transpose a Matrix?

A = [
    [1, 2, 3],
    [4, 5, 6]
]

# Create an empty transpose matrix
transpose = [
    [0, 0],
    [0, 0],
    [0, 0]
]

# Transpose logic
for i in range(2):        # rows of A
    for j in range(3):    # columns of A
        transpose[j][i] = A[i][j]

# Print transpose
print("Transpose Matrix:")
for row in transpose:
    print(*row)
```

```python
# Write a Python Program to Sort Words in Alphabetic Order?

# take input from user
sentence = input("Enter the sentence :")
#split sentence into words
words = sentence.split()
#sorts the words alphabetically
words.sort()
#print the sortd words
print("Sorted words :")
for word in words:
    print(word)
```

```python
# Write a Python Program to Remove Punctuation From a String?

punctuations = "!()-[]{};:'\"\\,<>./?@#$%^&*_~"

string = input("Enter a string: ")

no_punc = ""

for char in string:
    if char not in punctuations:
        no_punc = no_punc + char

print("String without punctuation:")
print(no_punc)
```
