1️⃣ Is the Python Standard Library included with PyInputPlus?

No.
PyInputPlus is not part of Python’s Standard Library. It is a third-party module that must be installed separately before you can use it.

2️⃣ Why is PyInputPlus commonly imported with import pyinputplus as pypi?

Because the full name “pyinputplus” is long, programmers use a shorter alias like “pypi” (or sometimes “pyip”) to save typing and make the code easier to read.
This is simply a naming convenience.

3️⃣ How do you distinguish between inputInt() and inputFloat()?

inputInt() is used when you want to accept only whole numbers (integers) as input.

inputFloat() is used when you want to accept decimal numbers (floating-point values) as well as integers.

In short:
inputInt() → whole numbers only.
inputFloat() → whole and decimal numbers.

4️⃣ Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?

By using PyInputPlus’s range-checking parameters.
You set a minimum value of 0 and a maximum value of 99.
This ensures that the user can only enter integers within that range — any value outside it will be rejected and the user will be asked again.

5️⃣ What is transferred to the keyword arguments allowRegexes and blockRegexes?

Both parameters take lists of regular expression (regex) patterns.

allowRegexes contains regexes that define inputs which should be explicitly allowed, even if they normally would not pass validation.

blockRegexes contains regexes that define inputs which should be explicitly rejected, even if they would normally be valid.

In short:
allowRegexes → patterns of inputs to accept.
blockRegexes → patterns of inputs to reject.

6️⃣ If a blank input is entered three times, what does inputStr(limit=3) do?

When limit=3 is set, the user gets three chances to enter valid input.
If the input is left blank (or invalid) all three times,
PyInputPlus raises a RetryLimitException — meaning the input attempts have been exhausted, and the program stops that input request.

7️⃣ If blank input is entered three times, what does inputStr(limit=3, default='hello') do?

In this case, the program still allows three invalid attempts, but after those attempts, instead of raising an exception,
it returns the default value “hello”.

So if the user fails to give valid input three times, the input variable is automatically set to “hello”.
