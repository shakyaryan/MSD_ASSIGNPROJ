# Question 1:
-  Define a class with a generator which can iterate the numbers, which are divisible by

 7, between a given range 0 and n

```python
def gennum(n):
    for i in range(n):
        if i % 7 == 0:
            yield i

for x in gennum(50):
    print(x)
```

Question 2:
Write a program to compute the frequency of the words from the input. The output
should output after sorting the key alphanumerically.
Suppose the following input is supplied to the program:
New to Python or choosing between Python 2 and Python 3? Read Python 2 or
Python 3.

```python
text = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."

words = text.split()

freq = {}

for w in words:
    freq[w] = freq.get(w, 0) + 1

for key in sorted(freq):
    print(key + ":" + str(freq[key]))
```

Question 3:

Define a class Person and its two child classes: Male and Female. All classes have a
method &quot;getGender&quot; which can print &quot;Male&quot; for Male class and &quot;Female&quot; for Female
class.

```python
class Person:
    def getGender(self):
        pass   # Parent has no specific gender

class Male(Person):
    def getGender(self):
        print("Male")

class Female(Person):
    def getGender(self):
        print("Female")

# Testing
m = Male()
m.getGender()   # Output: Male

f = Female()
f.getGender()   # Output: Female
```

Question 6:
Please write a binary search function which searches an item in a sorted list. The
function should return the index of element to be searched in the list.

```python
#4

subjects = ["I", "You"]
verbs = ["Play", "Love"]
objects = ["Hockey", "Football"]

sentences = [f"{s} {v} {o}" for s in subjects for v in verbs for o in objects]

for sentence in sentences:
    print(sentence)
```

```python
#5
import zlib

text = "hello world!hello world!hello world!hello world!"

# compress
compressed = zlib.compress(text.encode())

# decompress
decompressed = zlib.decompress(compressed).decode()

print("Original:", text)
print("Compressed:", compressed)
print("Decompressed:", decompressed)
```

```python
#6
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid           # found, return index
        elif arr[mid] < target:
            low = mid + 1       # search right half
        else:
            high = mid - 1      # search left half

    return -1                  # not found
```

```python

arr = [1, 3, 5, 7, 9, 11]
print(binary_search(arr, 7))   # prints 3
print(binary_search(arr, 2))   # prints -1

```
