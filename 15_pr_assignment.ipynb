Question 5:
Define a class named Shape and its subclass Square. The Square class has an init function
which takes a length as argument. Both classes have a area function which can print the area
of the shape where Shape&#39;s area is 0 by default.

```python
# sol 1
def gen_numbers(n):
    for i in range(n+1):
        if i % 35 == 0:   # divisible by both 5 and 7
            yield i

n = int(input())
print(",".join(str(i) for i in gen_numbers(n)))
```

```python
# sol2
def even_gen(n):
    for i in range(0, n+1, 2):
        yield i

n = int(input())
print(",".join(str(i) for i in even_gen(n)))
```

```python
# sol3
n = int(input())

fib = [0, 1]
[fib.append(fib[-1] + fib[-2]) for _ in range(2, n+1)]

print(",".join(str(i) for i in fib[:n+1]))
```

```python
#sol5
class Shape:
    def area(self):
        return 0

class Square(Shape):
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length * self.length

s = Square(5)
print(s.area())   # Example
```
