1. Create a list called years_list, starting with the year of your birth, and each year thereafter until the year of your fifth birthday.

Answer:

Use a variable `birth_year` and a list comprehension or `range()`; for example with `birth_year = 1990`:

```python
birth_year = 1990
years_list = list(range(birth_year, birth_year + 6))
# years_list == [1990, 1991, 1992, 1993, 1994, 1995]
```

Answer:

The year of your third birthday is `birth_year + 3`. With the example above: `1993`.

```python
third_birthday_year = birth_year + 3  # 1993 for birth_year = 1990
```

Answer:

6. Make a surprise list with the elements "Groucho", "Chico", and "Harpo".

Answer:

```python
surprise = ["Groucho", "Chico", "Harpo"]
```

---

7. Lowercase the last element of the surprise list, reverse it, and then capitalize it.

Answer:

Last element `"Harpo"` lowercased -> `"harpo"`, reversed -> `"oprah"`, capitalized -> `"Oprah"`. If you want to replace the list element with that result:

```python
s = surprise[-1].lower()[::-1].capitalize()
surprise[-1] = s
print(surprise)  # ['Groucho', 'Chico', 'Oprah']
```

8. Make an English-to-French dictionary called e2f and print it. Here are your starter words: dog is chien, cat is chat, and walrus is morse.

Answer:

```python
e2f = {'dog': 'chien', 'cat': 'chat', 'walrus': 'morse'}
print(e2f)  # {'dog': 'chien', 'cat': 'chat', 'walrus': 'morse'}
```

---

9. Write the French word for walrus in your three-word dictionary e2f.

Answer:

Access it with:

```python
french_walrus = e2f['walrus']  # 'morse'
```

---

10. Make a French-to-English dictionary called f2e from e2f. Use the items method.

Answer:

Swap keys and values using `.items()`:

```python
f2e = {french: english for english, french in e2f.items()}
# f2e == {'chien': 'dog', 'chat': 'cat', 'morse': 'walrus'}
```

---

11. Print the English version of the French word chien using f2e.

Answer:

```python
print(f2e['chien'])  # 'dog'
```

12. Make and print a set of English words from the keys in e2f.

Answer:

```python
english_words = set(e2f.keys())
print(english_words)  # {'dog', 'cat', 'walrus'}
```

13. Make a multilevel dictionary called life. Use these strings for the topmost keys: 'animals', 'plants', and 'other'. Make the 'animals' key refer to another dictionary with the keys 'cats', 'octopi', and 'emus'. Make the 'cats' key refer to a list of strings with the values 'Henri', 'Grumpy', and 'Lucy'. Make all the other keys refer to empty dictionaries.

Answer:

```python
life = {
    'animals': {
        'cats': ['Henri', 'Grumpy', 'Lucy'],
        'octopi': {},
        'emus': {}
    },
    'plants': {},
    'other': {}
}
```

---

14. Print the top-level keys of life.

Answer:

```python
print(list(life.keys()))  # ['animals', 'plants', 'other']
```

---

15. Print the keys for life['animals'].

Answer:

```python
print(list(life['animals'].keys()))  # ['cats', 'octopi', 'emus']
```

---

16. Print the values for life['animals']['cats']

Answer:

```python
print(life['animals']['cats'])  # ['Henri', 'Grumpy', 'Lucy']
```
